```python
import requests
from bs4 import BeautifulSoup

r=requests.get('https://www.century21.com/real-estate/rock-springs-wy/LCWYROCKSPRINGS/')
c=r.content


soup=BeautifulSoup(c, 'html.parser')


all=soup.find_all('div',{'class':'property-card'})

all[0].find('a', {'class':'listing-price'}).text.replace('\n', '').strip()
```

```python
for item in all:
    print(item.find('a', {'class':'listing-price'}).text.replace('\n', '').strip())
    print(item.find('div',{'class':'property-address'}).text.replace('\n', '').strip())
    print(item.find('div',{'class':'property-city'}).text.replace('\n', '').strip())
    try:
        print('Beds ' + item.find('div',{'class':'property-beds'}).text.replace('\n', '').strip())
        print('Baths ' + item.find('div',{'class':'property-baths'}).text.replace('\n', '').strip())
    except:
        print('No info given for Beds/Baths')
        pass
    try:
        print('Half-Baths ' + item.find('div',{'class':'property-half-baths'}).text.replace('\n', '').strip())
    except:
        pass
    try:
        print(item.find('div',{'class':'property-sqft'}).text.replace('\n', '').strip())
    except:
        print('No info given for sq ft.')
        pass
    print('')
```
